#Risk vs Returns graph

Let's pull some of the oil & gas company stocks using 'web.DataReader'


```python
import datetime
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib as mpl

mpl.rc('figure', figsize=(10,10))
mpl.__version__

style.use('ggplot')

start = datetime.datetime(2020, 1, 1)
end   = datetime.datetime(2020, 7, 1)

mygroup = ['DVN', 'APA', 'EOG', 'COP', 'MRO', 'NBL', 'PXD', 'COG', 'PXD','EQT','DNR','FANG','XOM','CVX', 'OXY','CXO']

dfcomp = web.DataReader(mygroup ,'yahoo',start=start,end=end)['Adj Close']
retscomp = dfcomp.pct_change()
```

Risk vs Returns

The average of returns (Return Rate) and the standard deviation of returns (Risk - std measures volatility in the market, or the average amount by which individual data points differ from the mean) are calculated.


```python
import matplotlib.cm as cm
import numpy as np


colors = cm.rainbow(np.linspace(0, 1, len(retscomp.std())))

for label,x,y,c in zip(retscomp.columns,retscomp.mean(),retscomp.std(),colors):
    plt.scatter(x,y, color = c, label = label)
    # to put label next to the point
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.legend()
plt.xlabel('Expected returns')
plt.ylabel('Risk')
plt.title('Risk vs Returns')
```
